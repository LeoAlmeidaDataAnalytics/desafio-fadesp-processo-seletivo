```python
# Importacao do pacote pandas
import pandas as pd 
```

```python
# Importacao dos dados e criacao do nome do dataframe ficticio
df = pd.read_csv('/content/dataset_desafio_fadesp.csv', encoding=('ISO-8859-1'))
df
```

```python
# Veremos assim as 10 primeiras linhas do dataframe, e conhecendo melhor o df.
df.head(10)
```

```python
# Obtendo melhores informacoes do dataframe, como numero de colunas, de linhas e se tem numeros nulos (NaN) ou nao.
df.info()
```

```python
# Fazendo uma copia do dataframe original com nome de df_pre
df_pre = df.copy()  
```

```python
# Formatando dados que originalmente estao como strings, mas na verdade sao datas.
df_pre['Order Date'] = pd.to_datetime(df_pre['Order Date'])
df_pre['Ship Date'] = pd.to_datetime(df_pre['Ship Date'])
```

```python
# Obtendo maiores informacoe do df pre-processado agora.
df_pre.info()
```

```python
# Comeco da limpeza de dados com remocao de valore nulos em todo o df pre-processado.
df_pre.dropna(inplace=True)
```

```python
# Confirmando a limpeza de dados.
df_pre.info()
```

```python
# Visualizando o df pre-processado.
df_pre
```

```python
# Removendo todas as linhas duplicadas em nosso dataframe.
df_pre.drop_duplicates()
```

```python
# Importando as bibliotecas Matplotib e Numpy
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Comecando aqui a visualizacao e analise dos dados.
df_pre.plot(x = 'State', y = 'Sales', title = 'Vendas por Estado')
```

```python
x = df_pre['Order Priority']
y = df_pre['Sales']
plt.title('Vendas por Prioridade de Entrega')
plt.bar(x,y)
```

```python
df_maiores_vendas = df_pre.sort_values(by='Sales', ascending=False).head(10).copy()
```

```python
x = df_maiores_vendas['City']
y = df_maiores_vendas['Sales']
```

```python
plt.figure(figsize=(15,8))
plt.title('Vendas nas 10 Maiores Cidades')
plt.bar(x,y)
```

```python
x = df_maiores_vendas['Sub-Category']
y = df_maiores_vendas['Sales']
```

```python
plt.title('Produtos mais Vendidos')
plt.bar(x, y)
```
